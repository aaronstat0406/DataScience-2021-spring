```python
import pandas as pd 
import numpy as np
```

```python
data=pd.read_csv("IMDB-Movie-Data.csv")
data_df=pd.DataFrame(data)
data_df
```

```python
#(1)
rating_2016=data_df.query("Year =='2016'")
top_3=rating_2016.nlargest(3,"Rating") #查詢top-3 movies

print("Top-3 movies with the highest ratings in 2016")
print(top_3["Title"])
```

```python
#(2)

df1=data_df["Revenue (Millions)"]
df1=df1.fillna(0) #把遺失的值換為0
df2=data_df["Actors"].str.split("|",expand=True)
df3=pd.concat([df2,df1],axis=1) #合併actors revenue的資料 (1000*5的df)
#print(type(df3))

#print(mix)
# 匯入 df3[0] df3[1] df3[2] df[3]

#print(type(df3[0]))
list3=df3[0].tolist()
#print(type(list3))

a={"Actors":df3[0],"Revenue":df1}
dfa=pd.DataFrame(a)
b={"Actors":df3[1],"Revenue":df1}
dfb=pd.DataFrame(b)
c={"Actors":df3[2],"Revenue":df1}
dfc=pd.DataFrame(c)
d={"Actors":df3[3],"Revenue":df1}
dfd=pd.DataFrame(d)

mix=pd.concat([dfa,dfb,dfc,dfd],ignore_index=True)
#print(mix)

final = mix.groupby("Actors").mean()
final=final.sort_values(by=["Revenue"],ascending=False)
#print(final)


top=final["Revenue"].max()

print("The actor generating the highest average revenue? ")
for i in range (len(final)):
    if final["Revenue"][i]==top: #top=936.63
        print(final.index[i])

#final_check=[]
#for i in range(len(final)):
   # final_check.append(final['Actors'][i])
#final_check
```

```python
#(3)
df4=data_df[["Rating"]]
df2=data_df[["Actors"]]
actors_rating=pd.concat([df2,df4],axis=1)
actors_rating


emma_rating=[]
for i in range(len(actors_rating)):
    for j in range(len(actors_rating["Actors"][i])):
        if "Emma Watson" in actors_rating["Actors"][i]:
            emma_rating.append(actors_rating["Rating"][i])
            break
ans=np.mean(emma_rating)  
ans="%.3f"%ans
    
#print(emma_rating)

print("The average rating of Emma Watson's movies")
print(ans)



```

```python
#不重複演員名單 actors_new
actors_list=data_df["Actors"].tolist()
for i in range(len(actors_list)):
    actors_list[i]=actors_list[i].replace("| ","|")
    actors_list[i]=actors_list[i].split("|")
#actors_list 二維list

actors_new=[]
for i in range(len(actors_list)):
    for j in range(len(actors_list[i])):
        if actors_list[i][j] not in actors_new:
            actors_new.append(actors_list[i][j])

actors_new
print(len(actors_new))
```

```python
#(4)

director=data_df["Director"]
director=director.unique().tolist() #.unique()不重複導演名單

co_actors=[]
for i in range(len(director)):
    temp=[] #存導演合作過不重複的演員
    for j in range(len(data_df)):
        if director[i]==data_df["Director"][j]:
            for k in range(len(actors_list[j])):
                if actors_list[j][k] not in temp:
                    temp.append(actors_list[j][k]) 
    co_actors.append(len(temp))
    
            
dict={"director":director,"co_actors_num":co_actors}
df=pd.DataFrame(dict)
df=df.sort_values(by="co_actors_num",ascending=False) 
df.head(4)
```

```python
#(5)

genre_fix=data_df["Genre"].tolist()
for i in range(len(genre_fix)):
    genre_fix[i]=genre_fix[i].replace("| ","|")
    genre_fix[i]=genre_fix[i].split("|") 
genre_fix

num_genre=[]
for i in range(len(actors_new)):
    save=[]
    for j in range(len(data_df)):
        if actors_new[i] in data_df["Actors"][j]:
            for k in range(len(genre_fix[j])):
                if genre_fix[j][k] not in save:
                    save.append(genre_fix[j][k])
    num_genre.append(len(save))
dict={"actor":actors_new,"num_genre":num_genre}
df=pd.DataFrame(dict)
df=df.sort_values(by="num_genre",ascending=False) 

top2=df["num_genre"][0:2].tolist()
top2

k=0
num_list=df["num_genre"].tolist() 
for i in range(len(top2)):
    while top2[i]==num_list[k]:  #找出num_genre的top2
        k+=1
df.head(k)
```

```python
#(6)

#actors_new 不重複演員名單
#actors_list 二維list

gap_year=[]

for i in range(len(actors_new)):
    year=[]
    for j in range(len(actors_list)):
        if actors_new[i] in actors_list[j]:
            year.append(data_df["Year"][j])
    if len(year)==0:
        gap_year.append(0)
    else:
        gap_year.append(max(year)-min(year))

dict={"actor":actors_new,"gap_year":gap_year}
df=pd.DataFrame(dict)
df=df.sort_values(by="gap_year",ascending=False) 
df

top3=df["gap_year"][0:3].tolist()
top3

k=0
num_list=df["gap_year"].tolist() 
for i in range(len(top3)):
    while top3[i]==num_list[k]:  #找出num_genre的top3
        k+=1
df=df.head(k)
df
```
